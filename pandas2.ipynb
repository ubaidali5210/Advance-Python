## .filter

```python
## 0 --> Rows, 1 ---> Columns
## Filter based on rows-wise operations
df.filter(items= [10,11], axis=0)

df.filter(items=[5,15], axis=0)

df.filter(items=['cgpa','placed'],axis=1)

# it is used to give you the filtered items 
# if you want it in rows you have to mention index no. in items
# if you want it in the column you have to mention name of the columns
```

## filter based on likeness

```python
## Filter based on likeness
df.filter(like ='6', axis=0)

df.filter(like= '3',axis=0)

# it is used to give you the no. which is present in the index 
```

## .where

```python
df = pd.read_csv('data\placement.csv')

## To where the data based on condition
df['cgpa'].where(df['cgpa'] > 7, other="Fail")

df['placed'].where(df['placed']>=1, other='zero')

# where is used to write conditions
```
