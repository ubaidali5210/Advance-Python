## arithmetic calculation in numpy

```python
max1=np.arange(2,20).reshape(3,6)
max2=np.arange(22,40).reshape(3,6)

# add
add = max1 + max2
print(add)

# subtract
sub = max1 - max2
print(sub)

# divide
div = max1 / max2
print(div)

# the order must be same of both of the matrix
```

## .dot multiply matrix

```python
max1=np.arange(2,20).reshape(3,6)
max2=np.arange(22,40).reshape(6,3)

multi = max1.dot(max2)
print(multi)

# column of 1st matrix must be equal to row of 2nd matrix
```
